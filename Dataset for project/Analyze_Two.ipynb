```python
df = df[['asin','brand','categories','description','price','salesRank','title']]
```

```python
df.isnull().sum()
```

```python
df2 = df.dropna()
print(len(df2))
```

```python
unique_categories = df2['categories'].unique()
unique_categories[0:25]
```

```python
import ast

unique_category_set = set()
for ele in unique_categories:
    temp = ast.literal_eval(ele)
    for ele2 in temp:
        if type(ele2) == list:
            for ele3 in ele2:
                unique_category_set.add(ele3)
        else:
            unique_category_set.add(ele2)
```

```python
print(unique_category_set)
print(len(unique_category_set))
```

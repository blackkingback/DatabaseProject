```python
raw_salesRank = df[['asin','salesRank']]
```

```python
f = open("asin_sales_rank.tsv","w")
f.write("asin"+"\t"+"sale_category"+"\t"+"sale_rank"+"\n")
for index,row in raw_salesRank.iterrows():
    asin = row['asin']
    salesRank = ast.literal_eval(row['salesRank'])
    if len(salesRank.keys()) != 1:
        temp_category = "Unknown Category"
        temp_salesRank = "-1"
        f.write(asin+"\t"+temp_category+"\t"+temp_salesRank+"\n")
    else:
        temp_category = list(salesRank.keys())[0]
        temp_salesRank = salesRank[temp_category]
        f.write(asin+"\t"+str(temp_category)+"\t"+str(temp_salesRank)+"\n")
f.close()
```
